```python
# Criando o banco de dados twitterdb
db = client.twitterdb
```

```python
# Criando a collection "col"
col = db.tweets
```

# An√°lise de Dados com Pandas e Scikit_learn

```python
# Criando um dataset com dados retornandos do MongoDB
dataset = [{"created_at":item["created_at"], "text":item["text"],} for item in col.find()]
```

```python
# importando o m√≥dulo Pandas para trabalhar com datasets em Python
import pandas as pd
```

```python
# Criando um data frame a partir do dataset
df = pd.DataFrame(dataset)
```

```python
# imprimindo o dataframe
df
```

```python
# importando o m√≥dulo Scikit Learn
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Contando o n√∫mero de ocorr√™ncia das principais palavras em nosso dataset
cv = CountVectorizer()
count_matrix = cv.fit_transform(df.text)
```

```python
# Contando o n√∫mero do ocorr√™ncia das principais palavras em nosso dataset
word_count = pd.DataFrame(cv.get_feature_names(), columns=["word"])
word_count["count"] = count_matrix.sum(axis=0).tolist()[0]
word_count = word_count.sort_values("count", ascending=False).reset_index(drop=True)
word_count[:50]
```
